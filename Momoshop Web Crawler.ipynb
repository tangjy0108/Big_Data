```python
import pandas as pd
from bs4 import BeautifulSoup
import json
import requests
import re
import time
from tqdm import tqdm
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36'}
# 請填10個搜尋詞彙
keywords = ['運動鞋']
pages = 5
urls = []
df = []
prdName = []
```

```python
for keyword in keywords:
    for page in range(1, pages):
      url = 'https://m.momoshop.com.tw/search.momo?_advFirst=N&_advCp=N&curPage={}&searchType=1&cateLevel=2&ent=k&searchKeyword={}&_advThreeHours=N&_isFuzzy=0&_imgSH=fourCardType'.format(page, keyword)
      print(url)
      resp = requests.get(url, headers=headers)
      if resp.status_code == 200:
        soup = BeautifulSoup(resp.text, features="lxml")
        #print(soup)  
            # 利用for迴圈得到畫面上所有商品的連結，並儲存到urls陣列裡面
        for item in soup.select('li.goodsItemLi > a'):
          print(item)
          content = "https://m.momoshop.com.tw"+item['href']
          product = soup.find_all('h3',{'class':'prdName'})
          urls.append(content)
      else:
        print("沒有資料")
        break
        #檢查是否有重複的urls並排除
        # urls = function(urls)
      for pro in product:
        prdName.append(pro.text)
```

```python

for i, url in enumerate(urls[:]):
  columns = []
  values = []

  resp = requests.get(url, headers=headers)
  soup = BeautifulSoup(resp.text, features="lxml")
  # 標題
  try:
   title = soup.find('meta', {'property': 'og:title'})['content']
  except:
    title = 'Nan'
  #print(title)
  # 品牌
  try:
    brand = soup.find('a',{'class','brandNameTxt'})['brandnameeng']
  except:
    brand = 'Nam'
  # 連結
  link = url
  #print(link)
  # 原價
  try:
    price = soup.find('del').text
    #price = soup.find('td',{'class','setpriceArea'}).text
    #price = re.sub('元',' ',price)
  except:
    price = 'Nan'
  # 特價
  amount = soup.find('p',{'class','priceTxtArea'}).text
  amount = re.sub('元',' ',amount)
  # 類型
  try:
    cate = ''.join([i.text for i in soup.findAll('article', {'class': 'pathArea'})])
    cate = re.sub('\n|\xa0', ' ', cate)
  except:
    cate = 'Nan'
  #print(cate)
  # 描述
  try:
    desc = soup.find('div', {'class': 'Area101'}).text
    desc = re.sub('\r|\n| ', '', desc)
  except:
    desc = 'Nan'

  
```

```python
for i in soup.select('div.attributesArea > table > tr'):
  try:
  # 整理規格的內容
  #column = i.find().text
    column = re.sub('\n|\r| ', '', column)
    value = ''.join([j.text for j in i.findAll('li')])
    value = re.sub('\n|\r| ', '', value)
    columns.append(column)
    values.append(value)
  except:
    pass
  ndf = pd.DataFrame(data=values, index=columns).T
  df.append(ndf)
df = pd.concat(df, ignore_index=True)
print(df.head)
```
